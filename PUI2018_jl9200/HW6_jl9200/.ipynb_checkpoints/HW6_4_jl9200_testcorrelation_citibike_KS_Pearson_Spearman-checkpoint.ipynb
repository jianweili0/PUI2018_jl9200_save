# Now retest using a test for correlation.

That will answer a slightly different question though - formulate the NULL appropriately. The tests for correlations (generally) requires the variable to be paired, so that I can tell if x changes does y change similarly. But the datasets are of different size! You will need to reduce them to the same size. You can do that by subsampling of the data: take only 1 ride every of 200, which you can achieve "slicing and broadcasting" the array or using one of the python function (built in python numpy.random.choice() functions for example: Docstring: choice(a, size=None, replace=True, p=None)

Generates a random sample from a given 1-D array

    .. versionadded:: 1.7.0
    
Parameters ...

But make sure you understand how to use it! there is an option "replace" which you should think about.

The scipy.stats KS test already tells me the significance and the p-value.

The next few cells are here just to show you how you would obtain the same result by hand, but they are not required.

Remember: the Null hypothesis is rejected if

$D_KS(n1,n2) >c(\alpha)\sqrt{\frac{(n1+n2)}{n1n2}} $

(see class notes) where $c(\alpha$) is the inverse of the KS distribution, and you do not have to know how to get that cause there are tables that list critical values!! 

http://www.real-statistics.com/tests-normality-and-symmetry/statistical-tests-normality-symmetry/kolmogorov-smirnov-test/kolmogorov-distribution/

But also this result depends in your choice of binning through, and thustheresultyou get by hand may not be exactly the same as the one the KS returns. Either way: this is how you would calculate the KS statistics by hand.

```python
# my alpha = 0.05, based on the k2sample table c(alpha) = 1.36



DS = lambda n1,n2: 1.36*np.sqrt((n1+n2)/(n1*n2))
# in CLASS NOTE UI5 PUI2018 PAGE 53/63, we need to compare the KS metric(which is statistic)


def printresult (n1,n2):
    ks = scipy.stats.ks_2samp(n1,n2)
    print(ks[0])
    print(DS(len(df.dayrider),len(df.nightrider)))
    if ks[0]>DS(len(n1),len(n2)):
        
        print("The null hypothesis is rejected")
    else:
        print("The null hypothesis is not rejected")

printresult(df.dayrider,df.nightrider)
```

# Pearson's test for correlation

notice that the Pearson's is a pairwise test: the samples need to be a. the same size b. sorted! (how??)

http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pearsonr.html#scipy.stats.pearsonr

```python
'''
You can do that by subsampling of the data: 
take only 1 ride every of 200, which you can 
achieve "slicing and broadcasting" the array 
or using one of the python function 
(built in python numpy.random.choice() functions 
for example: Docstring: choice(a, size=None, replace=True, p=None)

'''

day_p = np.random.choice(df.dayrider,size =200,replace = False,p = None)
night_p = np.random.choice(df.nightrider,size = 200, replace = False,p = None)

day_p.sort()
night_p.sort()

Pearson_test = scipy.stats.pearsonr(day_p,night_p)
print(Pearson_test)
```

The Pearson correlation coefficient is between -1 to 1, the result of Pearson_test is 0.93 which is close to 1 will indicate that they are significant positive correlation between the two dataset. The p-value is very small around 0, which mean that this test indicate the probability of an uncorrelated system producing datasets. which mean this test is reliable. and P-value < 0.05 so we can reject the null hypothesis



# Spearman's test for correlation

http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.spearmanr.html#scipy.stats.spearmanr

```python
Spearman_test = scipy.stats.spearmanr(day_p,night_p)
print(Spearman_test)
```
